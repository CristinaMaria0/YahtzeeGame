*Problemă: Yahtzee*

Jocul presupune aruncatul simultan a doar 5 zaruri
în scopul obtinerii a unor combinatii de numere ale zarurilor aruncate:


*   Yahtzee - 5 zaruri cu aceeasi valoare
*   Four of a kind - 4 zaruri cu aceeasi valoare
*   Three of a kind - 3 zaruri cu aceeasi valoare
*   Full House  
*   Large Straight - 5 zaruri cu valori consecutive
*   Small Straight - 4 zaruri cu valori consecutive

Am calculat probabilitatiile de aparitie pentru fiecare caz in urma a N (imput de la tastatura) aruncari de zaruri si am plotat un grafic cu aceste probabilitati pentru a putea vizualiza mai bine rezultatele finale.



```python

import random
import matplotlib.pyplot as plt
from collections import Counter
import numpy as np
```

```python
def consecutive(l):
  return sorted(l) == list(range(min(l), max(l)+1))

def yahtzee_problem(num_incercari):
    num_yahtzee = 0
    num_full_house=0
    num_four_kind=0
    num_three_kind=0
    num_large_straight=0
    num_small_straight=0
    #aruncam zarurile de num_incercari ori
    for i in range(num_incercari):
        zar = []
        #completam vectorul zar cu numere random de la 1 la 6
        for j in range(5):
            zar.append(random.randrange(1, 7))

        counts = [zar.count(i) for i in range(1, 7)]
        #retinem combinatiile de zaruri aparute
        if 5 in counts:
          num_yahtzee+=1
        elif sorted(Counter(zar).values()) == [2, 3]:
          num_full_house+=1
        elif 4 in counts:
          num_four_kind+=1
        elif 3 in counts:
          num_three_kind+=1
        elif consecutive(zar) and len(set(zar)) == 5:
           num_large_straight+=1
        elif consecutive(set(zar)) and len(set(zar))==4:
          num_small_straight+=1


    # Calculam ratele pentru fiecare posibilitate
    rate_yahtzee = num_yahtzee / num_incercari
    rate_full_house = num_full_house / num_incercari
    rate_four_kind = num_four_kind / num_incercari
    rate_three_kind = num_three_kind / num_incercari
    rate_large_straight = num_large_straight / num_incercari
    rate_small_straight = num_small_straight / num_incercari

    # Afisam probabilitatile fiecarui caz
    print(f"Rata Yahtzee: {rate_yahtzee}; numarul de Yahtzee a fost: {num_yahtzee}")
    print("Rata Full House: ", rate_full_house)
    print("Rata Four of a Kind: ", rate_four_kind)
    print("Rata Three of a Kind: ", rate_three_kind)
    print("Rata Large Straight: ", rate_large_straight)
    print("Rata Small Straight: ", rate_small_straight)

    # Afisam totul intr-un grafic pentru a vedea mai bine rezultatele
    labels = ['Yahtzee', 'Full House', 'Four of a Kind', 'Three of a Kind', 'Large Straight', 'Small Straight']
    sizes = [num_yahtzee, num_full_house, num_four_kind, num_three_kind, num_large_straight, num_small_straight]
    explode_yahtzee=[0.1,0,0,0,0,0] #punem in evidenta probabilitatea unui Yehtzee
    plt.pie(sizes, labels=labels, explode=explode_yahtzee, startangle=90, autopct='%.2f%%')
    plt.title('Proportia rezultatelor în ' + str(num_incercari) + ' de incercari')
    plt.show()
```

```python
#introducem de la tastatură numărul de aruncări pe care îl dorim
number_of_trials = int(input("Dati numarul de incercari: "))
yahtzee_problem(number_of_trials)
```
